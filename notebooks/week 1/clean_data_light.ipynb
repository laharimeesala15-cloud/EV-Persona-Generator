```python
import pandas as pd

df = pd.read_csv("../data/electric_vehicles_spec_2025.csv.csv")

print("âœ… Dataset Loaded Successfully!")
print("ðŸ”¹ First 5 rows:")
print(df.head())

print("\nðŸ”¹ Columns available:")
print(df.columns.tolist())

print("\nðŸ”¹ Missing values summary:")
print(df.isnull().sum())
```

```python
# Drop columns that aren't useful for your EV persona generator
df_clean = df.drop(columns=['source_url'], errors='ignore')

# Drop duplicate rows (if any)
df_clean = df_clean.drop_duplicates()

print("âœ… Removed extra columns and duplicates")
print("Shape after cleaning:", df_clean.shape)
```

```python
df_clean.rename(columns={
    'top_speed_kmh': 'Top_Speed',
    'battery_capacity_kWh': 'Battery_kWh',
    'efficiency_wh_per_km': 'Efficiency',
    'range_km': 'Range_km',
    'acceleration_0_100_s': 'Accel_0_100',
    'cargo_volume_l': 'Cargo_L',
    'towing_capacity_kg': 'Towing_KG',
}, inplace=True)

print("âœ… Columns renamed!")
print(df_clean.columns.tolist())
```

```python
df_clean.to_csv("../data/ev_data_cleaned.csv", index=False)
print("ðŸ’¾ Cleaned data saved successfully as ev_data_cleaned.csv!")
```
